```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('customer_shopping_behavior.csv')
```

```python
df.head()
```

```python
#generate descriptive statistics info. of the dataset, using include all for the non numeric datas  
df.describe(include='all')
```

```python
#checking for null values and if found then '.sum' to get the idea of nul values
df.isnull().sum()
```

```python
#incase of missing values the best pactice is to perform imputation of median values of the column in place of missing values but here the 
#reviews belong to different categories so imputation with a common value for multiple categories doesnt make sense so will seek help of 
#lambda function to dedicate different median reviews for the respective categoties.

df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x:x.fillna(x.median()))
```

```python
#checking whether null values have been treated or not
df.isnull().sum()
```

```python
#formatting the column names to a standard format for easier handling 

df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
```

```python
df.head(20)
```

```python
#changing the column name to easier name

df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
#1st feature engineering: perform binning on the range of ages to separate categories using qcut(q=4) function

z=['young','adult','mid_aged','senior']
df['age_group']=pd.qcut(df['age'],q=4,labels=z)
```

```python
df.columns
```

```python
df[['age','age_group']].head(100)
```

```python
#2nd feature engineering: adding a new column to specifiy the purchase_frequencies_days by using dictionary allocating the categorical data to numbers using .map()

frequency_dict={
    'Fortnightly':14,
    'Weekly':7,
    'Monthly':30,
    'Quarterly':90,
    'Bi-Weekly':14,
    'Annually':365,
    'Every 3 Months':90}


df['purchase_frequency_days']=df['frequency_of_purchases'].map(frequency_dict)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
df
```

```python
df[['discount_applied','promo_code_used']].head()
```

```python
#identifying the redundancy as it is observed we have both columns named discount_applied & promo_code_used with identical records
#by using'==' for comparison and .all() for getting a one line result from a data where there's thousands of rows

df['discount_applied']==df['promo_code_used']
```

```python
(df['discount_applied']==df['promo_code_used']).all()
```

```python
#removing the redundant column 'promo_code_used' as we already have a column named 'discount_applied' with identical records

df.drop('promo_code_used',axis=1)
```
